```
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    'years': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'total_forest_cover_million': [70120, 70230, 70350, 70450, 70560, 70610, 70700, 70750, 70810, 70900, 70950, 71000, 71100],
    'deforestation': [200, 180, 150, 200, 250, 220, 180, 200, 150, 180, 200, 220, 180]
}

df = pd.DataFrame(data)

plt.bar(df['years'], df['deforestation'], color='g')
plt.title("Deforestation Analysis in India (2012–2024) in km²")
plt.xlabel("Years")
plt.ylabel("Deforestation (in km²)")
plt.xticks(rotation=45)
plt.show()
```
